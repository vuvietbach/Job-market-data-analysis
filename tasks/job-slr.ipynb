```python
import pandas as pd
import re

def update_salary(row):
    salary_str = row['salary']
    if not isinstance(salary_str, str):
        return salary_str
    matches = re.findall(r'([\d,\.]+)', salary_str)
    if len(matches) == 0:
        description = row['description']
        if isinstance(description, str):
            match = re.search(r'lương[^a-zA-Z0-9]*(\S+\s+){0,4}(tr|triệu|đồng|VND|USD|\$)', description)
            if match:
                if re.search(r'\b(tr|triệu|đồng|VND|USD|\$)\b', match.group(), flags=re.IGNORECASE):
                    result = re.sub(r'[,:.](?=\D|$)', '', match.group())
                    return result
    return row['salary']

df['salary'] = df.apply(update_salary, axis=1)
```

```python
def extract_salary_range(salary_str):
    if not isinstance(salary_str, str):
        return [0, 0]
    matches = re.findall(r'([\d,\.]+)', salary_str)
    if len(matches) == 2:
        if len(matches[0])>8:
            matches[0] = matches[0][:-4]
        if len(matches[1])>8:
            matches[1] = matches[1][:-4]
        if '$' in salary_str or 'USD' in salary_str:
            matches[0] = matches[0].replace(',', '')
            number1 = float(matches[0].replace('.', ''))
            number1 = round(number1*0.02426, 2)
            matches[1] = matches[1].replace(',', '')
            number2 = float(matches[1].replace('.', ''))
            number2 = round(number2*0.02426, 2)
        else:
            number1 = float(matches[0].replace(',', '.'))
            number2 = float(matches[1].replace(',', '.'))
        return [number1, number2]

    if len(matches) == 1:
        if len(matches[0])>8:
            matches[0] = matches[0][:-4]
        if '$' in salary_str or 'USD' in salary_str:
            matches[0] = matches[0].replace(',', '')
            number = float(matches[0].replace('.', ''))
            number = round(number*0.02426, 2)
        else:
            number = float(matches[0].replace(',', '.'))
        if "Lên tới" in salary_str or "Tới" in salary_str or "Đến" in salary_str or "Up to" in salary_str:
            return [0, number]
        else:
            return [number, 0]

    return [0, 0]
```

## EDA: Trực quan hóa dữ liệu
**1. Nhu cầu tuyển dụng theo mức lương**

Hiện tại đang phân thành các khoảng: Dưới 5 triệu, Từ 5-10 triệu, Từ 10-20 triệu, Từ 20-30 triệu, Từ 30-50 triệu và Trên 50 triệu, Thỏa thuận. 
* Đối với những dữ liệu khoảng (đầy đủ min_salary, max_salary), thực hiện lấy trung bình cộng, sau đó cho vào khoảng tương ứng. 
* Đối với dữ liệu thiếu 1 (thiếu min_salary hoặc max_salary), lấy giá trị còn lại làm giá trị đại diện. 
* Đối với dữ liệu thiếu 2, thì xếp vào nhóm Thỏa thuận

```python
df['average_salary'] = df.apply(lambda row: (row['min_salary'] + row['max_salary']) / 2
                                  if row['min_salary'] > 0 and row['max_salary'] > 0
                                  else (row['min_salary'] + row['max_salary'])
                                  if row['min_salary'] == 0 and row['max_salary'] > 0
                                  else (row['min_salary'] + row['max_salary'])
                                  if row['min_salary'] > 0 and row['max_salary'] == 0
                                  else 0, axis=1)

def categorize_salary_range(average_salary):
    if 0 < average_salary < 5:
        return 'Dưới 5 triệu'
    elif 5 <= average_salary < 10:
        return 'Từ 5-10 triệu'
    elif 10 <= average_salary < 20:
        return 'Từ 10-20 triệu'
    elif 20 <= average_salary < 30:
        return 'Từ 20-30 triệu'
    elif 30 <= average_salary < 50:
        return 'Từ 30-50 triệu'
    elif average_salary >= 50:
        return 'Trên 50 triệu'
    else:
        return 'Thỏa thuận'

df['salary_range'] = df['average_salary'].apply(categorize_salary_range)
salary_order = ["Dưới 5 triệu", "Từ 5-10 triệu", "Từ 10-20 triệu", "Từ 20-30 triệu", "Từ 30-50 triệu", "Trên 50 triệu", "Thỏa thuận"]
salary_counts = df['salary_range'].value_counts().reindex(salary_order).fillna(0)

plt.figure(figsize=(10, 6))
barplot = sns.barplot(x=salary_counts.index, y=salary_counts.values, palette="viridis", order=salary_order)
for idx, value in enumerate(salary_counts.values):
    barplot.text(idx, value + 0.1, str(value), ha='center', va='bottom', fontsize=8)

plt.xticks([])
plt.title("Nhu cầu tuyển dụng theo mức lương", fontweight='bold', color = 'blue')
plt.xlabel("Mức lương")
plt.ylabel("Số lượng công việc")

legend_labels = sns.color_palette("viridis", len(salary_counts))
legend_handles = [plt.Rectangle((0, 0), 1, 1, color=label) for label in legend_labels]
plt.legend(legend_handles, salary_counts.index, title="Mức lương")

plt.show()
```

**4. Thống kê số lượng công việc theo chế độ làm việc: Full-time, part-time, linh hoạt**

Tiến hành phân loại trên **3298 công việc**, trong đó có **2475 công việc có thông tin** và 823 công việc không cung cấp đủ thông tin về chế độ làm việc: **full-time, part-time, linh hoạt.**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def categorize_workmode(mode):
    if mode == "Toàn thời gian" or mode == "Toàn thời gian cố định":
        return 'Full-time'
    elif mode == "Linh hoạt":
        return 'Linh hoạt'
    elif mode == "Bán thời gian" or mode == "Bán Thời gian":
        return 'Part-time'
    elif mode == "Hợp đồng" or mode == "Theo hợp đồng tư vấn":
        return 'Full-time'
    elif mode == "Thực tập" or mode == "Làm từ xa":
        return 'Part-time'
    elif mode == "Tại văn phòng":
        return 'Full-time'
    else:
        return 'Không rõ/không đủ thông tin'

df['mode'] = df['work_mode'].apply(categorize_workmode)
mode_order = ["Full-time", "Part-time", "Linh hoạt", "Không rõ/không đủ thông tin"]
mode_counts = df['mode'].value_counts().reindex(mode_order).fillna(0)

plt.figure(figsize=(10, 6))
barplot = sns.barplot(x=mode_counts.index, y=mode_counts.values, palette="viridis", order=mode_order)
for idx, value in enumerate(mode_counts.values):
    barplot.text(idx, value + 0.1, str(value), ha='center', va='bottom', fontsize=8)

plt.title("Thống kê số lượng công việc theo chế độ làm việc", fontweight='bold', color = 'blue')
plt.xlabel("Chế độ làm việc")
plt.ylabel("Số lượng công việc")
plt.show()
```

**5. Biểu đồ thống kê số lượng công việc theo category**

Hàm **plot_salary(list: category)** dưới đây cho phép hiển thị biểu đồ thống kê số lượng công việc theo category được mô tả trong list.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import json
# # Đường dẫn file
# file_data = '/kaggle/input/1357913579/it_job_exp_norm_numberize.jsonl'
# with open(file_data, 'r') as file:
#     data_list = []
#     for line in file:
#         data = json.loads(line)
#         data_list.append(data)
#     df = pd.DataFrame(data_list)
def plot_salary(list_category):
    result_df = pd.DataFrame(columns=['category', 'count'])
    
    def count_jobs(df, category):
        exp_df = df.explode('category')
        count = exp_df[exp_df['category'] == category].shape[0]
        return count
    
    for i in range(len(list_category)):
        count_job = count_jobs(df, list_category[i])
        result_df = pd.concat([result_df, pd.DataFrame({'category': list_category[i], 'count': [count_job]})], ignore_index=True)

    labels = list_category
    result_df['label'] = labels

    palette = sns.color_palette("viridis", n_colors=len(labels))

    fig, ax = plt.subplots(figsize=(12, 6))
    bars = sns.barplot(x='label', y='count', data=result_df, palette=palette)
    for bar, label in zip(bars.patches, result_df['count']):
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width() / 2, height, f'{label}', ha='center', va='bottom', fontsize=10)

    legend_data = {'Nhãn phân loại': labels, 'Màu': palette}
    legend_df = pd.DataFrame(legend_data)
    legend_df = legend_df[legend_df['Nhãn phân loại'].isin(result_df['category'])] 

    handles = [plt.Line2D([0], [0], marker='o', color='w', markerfacecolor=color, markersize=12) for color in palette]

    legend = plt.legend(handles=handles, labels=labels, title='Category', bbox_to_anchor=(1.05, 1), loc='upper left')
    legend.get_title().set_fontweight('bold')
    legend.get_title().set_fontsize('12')
    
    if len(labels)>5 and len(labels)<8:
        plt.xticks(rotation=45, ha='right')
    elif len(labels)<12:
        plt.xticks(rotation=90, ha='right')
    elif len(labels)>=12:
        plt.xticks([])
        
    plt.xlabel('Category')
    plt.ylabel('Số lượng công việc')
    plt.title(f'Thống kê số lượng công việc theo category', fontsize=16, fontweight='bold', color='blue')
    plt.grid(axis='y')
    plt.show()
```

```python
list_category = ['Thiết kế đồ họa','Xây dựng','Kiến trúc','Chăm sóc khách hàng','Khoa học - Kỹ thuật','Bán hàng - Kinh doanh','IT Phần cứng - mạng','IT phần mềm','Copywriter', 'IT Phần cứng / Mạng']
plot_salary(list_category)
```

**6. Xây dựng hàm lọc các công việc theo lương**

**Ghi chú to đùng:** Nhớ lưu dữ liệu xử lý về lương thành một file .jsonl mới trước khi xử lý.

Nếu không thì cần chạy lại code xử lý lương ở trên (thì mới có trường min_salary và max_salary)

Dưới đây, hàm **filter_jobs_by_salary(aug1, aug2 = 0)** sẽ trả về một DataFrame của các dòng trong DataFrame gốc thỏa mãn điều kiện đưa vào. 
* **aug1** là dấu so sánh, có 4 trường hợp: bằng, lớn hơn, nhỏ hơn, **"không"** (code: '=', '>', '<', '0')
* **aug2** là số cần so sánh (đơn vị: triệu đồng). **Đối với code '0':** **"không"**, **aug2** sẽ để mặc định là 0. Mục đích của phép so sánh "không" này để đưa ra các dòng không có thông tin lương/thỏa thuận về lương.

Việc đưa ra DF sẽ dễ tích hợp nếu như cần lọc nhiều điều kiện, ví dụ lọc theo nhãn với mức lương thỏa mãn điều kiện cho trước ở khu vực truy vấn.

**Cách truy vấn:**
* Nếu là bằng, đưa ra các dòng có giá trị truy vấn nằm trong khoảng từ min_salary đến max_salary.
* Nếu là lớn hơn, đưa ra các dòng có giá trị truy vấn lớn hơn hoặc bằng trung bình cộng của min_salary và max_salary
* Nếu là bé hơn, đưa ra các dòng có giá trị truy vấn bé hơn hoặc bằng trung bình cộng của min_salary và max_salary (đương nhiên là ít nhất phải tồn tại giá trị min_salary hoặc max_salary để tránh trường hợp về 0)
* Nếu là "không", đưa ra các dòng có min_salary và max_salary đều bằng 0.

```python
def filter_jobs_by_salary(code, salary_q = 0):
    if code == '0':  # "không"
        filter_df = df[(df['min_salary'] == 0) & (df['max_salary'] == 0)]
    else:
        avg_salary = (df['min_salary'] + df['max_salary']) / 2
        if code == '=':  # bằng
            filter_df = df[(df['min_salary'] <= salary_q) & (df['max_salary'] >= salary_q)]
        elif code == '>':  # lớn hơn
            filter_df = df[avg_salary >= salary_q]
        elif code == '<':  # bé hơn
            filter_df = df[(avg_salary <= salary_q) & ((df['max_salary'] > 0) | (df['min_salary'] > 0))]
        else:
            raise ValueError("Giá trị aug1 không hợp lệ!")

    return filter_df 
```

```python
def extract_province(city):
    for i in range(len(region_name)):
        if region_name[i].lower() in city.lower():

            if i < 63:
                return region_name[i+ 63]
            else:
                if region_name[i]=='TP.HCM':
                    return 'Hồ Chí Minh'
                if region_name[i]=='Ha Noi':
                    return 'Hà Nội'
                if region_name[i]=='Thanh Hoá':
                    return 'Thanh Hóa'
                if region_name[i]=='Bà Rịa-Vũng Tàu':
                    return 'Bà Rịa - Vũng Tàu'
                if region_name[i]=='Nước Ngoài':
                    return 'Quốc tế'
                return region_name[i]
    return 'Other'

df['province'] = df['location'].apply(extract_province)
```
